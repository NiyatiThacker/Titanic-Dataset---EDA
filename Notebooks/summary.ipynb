```python
import pandas as pd

df = pd.read_csv("Titanic_cleaned_no_outliers.csv")


print(df.info())

print(df.describe(include='all'))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = ['Age', 'Fare', 'SibSp']

for col in numeric_cols:
    # Histogram
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()

    # Boxplot
    plt.figure(figsize=(6, 1))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation for numeric columns
corr = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting relevant numeric and categorical features
cols = ['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']

# Pairplot with hue set to Survived
sns.pairplot(df[cols], hue='Survived', diag_kind='kde', palette='Set2')
plt.suptitle('Pairplot of Titanic Features', y=1.02)
plt.show()
```
