```python
import pandas as pd
df = pd.read_csv("Titanic_cleaned_no_outliers.csv")
from scipy.stats import zscore

# Calculate Z-scores for Fare
df['Fare_z'] = zscore(df['Fare'])

# Display outliers: Z > 3 or Z < -3
fare_outliers = df[df['Fare_z'].abs() > 3]
print(f"Fare outliers found: {len(fare_outliers)}")
fare_outliers[['Fare', 'Fare_z']]
```

```python
# Calculate Z-scores for Age
df['Age_z'] = zscore(df['Age'])

# Display outliers: Z > 3 or Z < -3
age_outliers = df[df['Age_z'].abs() > 3]
print(f"Age outliers found: {len(age_outliers)}")
age_outliers[['Age', 'Age_z']]
```
